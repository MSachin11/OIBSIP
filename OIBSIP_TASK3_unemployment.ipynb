```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Unemployment in India.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
print(df.columns)
```

```python
df[' Date'] = pd.to_datetime(df[' Date'])
```

```python
plt.figure(figsize=(14, 7))
sns.lineplot(x=' Date', y=' Estimated Unemployment Rate (%)', data=df)
plt.title(' Estimated Unemployment Rate Over Time')
plt.xlabel(' Date')
plt.ylabel(' Estimated Unemployment Rate (%)')
plt.grid(True)
plt.show()
```

```python
print("Basic Statistics:")
print(df[' Estimated Unemployment Rate (%)'].describe())
```

```python
covid_start_date = '2020-03-01'
df_covid = df[df[' Date'] >= covid_start_date]

plt.figure(figsize=(14, 7))
sns.lineplot(x=' Date', y=' Estimated Unemployment Rate (%)', data=df_covid)
plt.title(' Estimated Unemployment Rate During Covid-19')
plt.xlabel(' Date')
plt.ylabel(' Estimated Unemployment Rate (%)')
plt.grid(True)
plt.show()
```

```python
print("Statistics During Covid-19:")
print(df_covid[' Estimated Unemployment Rate (%)'].describe())
```

```python
peak_unemployment = df_covid[df_covid[' Estimated Unemployment Rate (%)'] == df_covid[' Estimated Unemployment Rate (%)'].max()]
print("Peak Unemployment Rate During Covid-19:")
print(peak_unemployment)
```
